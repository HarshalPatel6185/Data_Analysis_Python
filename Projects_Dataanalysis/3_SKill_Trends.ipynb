```python
# Importing Libraries
import ast
import pandas as pd
import seaborn as sns
from datasets import load_dataset
import matplotlib.pyplot as plt  

# Loading Data
dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

# Data Cleanup
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
df['job_skills'] = df['job_skills'].apply(lambda x: ast.literal_eval(x) if pd.notna(x) else x)
```

```python
df
```

```python
df_dataanalyst_India = df[(df['job_country'] =='India') & (df['job_title_short'] == 'Data Analyst')]
df_dataanalyst_India['job_posted_month_no'] = df_dataanalyst_India['job_posted_date'].dt.month
df_dataanalyst_India_explode = df_dataanalyst_India.explode('job_skills')
df_dataanalyst_India_explode
```

```python
df_dataanalyst_India
```

```python
df_dataanalyst_India = df_dataanalyst_India.explode('job_skills')
```

```python
df_dataanalyst_India = df_dataanalyst_India.pivot_table(index = 'job_posted_month_no' , columns = 'job_skills', aggfunc = 'size').fillna(0)
df_dataanalyst_India
```

```python
df_dataanalyst_India.loc['Total'] = df_dataanalyst_India.sum()
df_dataanalyst_India = df_dataanalyst_India[df_dataanalyst_India.loc['Total'].sort_values(ascending=False).index]
df_dataanalyst_India
```

```python
df_dataanalyst_India = df_dataanalyst_India.drop('Total')
df_dataanalyst_India_plot = df_dataanalyst_India.iloc[: , :5]
df_dataanalyst_India
```

```python
df_dataanalyst_India_plot.plot(kind = 'line')
```

```python
df_month_counts = df[(df['job_country'] =='India') & (df['job_title_short'] == 'Data Analyst')]
df_month_counts['job_posted_month_no'] = df_month_counts['job_posted_date'].dt.month
#df_month_counts = df_month_counts.explode('job_skills')
df_month_counts
```

```python
df_month_counts = df_month_counts['job_posted_month_no'].value_counts().reset_index(name = 'T_skills_permonth')
```

```python
df_merged = pd.merge(df_dataanalyst_India ,df_month_counts,on='job_posted_month_no' )
```

```python
df_merged
```

```python

df_percent = df_merged.div(df_merged['T_skills_permonth']/100,axis=0 )
df_percent
# skill_cols = df_merged.columns.drop(['job_posted_month_no', 'T_skills_permonth'])

# df_percent = df_merged[skill_cols] \
#     .div(df_merged['T_skills_permonth'], axis=0) * 100

# df_percent['job_posted_month_no'] = df_merged['job_posted_month_no']


df_percent['job_posted_month_no'] = df_merged['job_posted_month_no']
```

```python
df_percent = df_percent.iloc[: , :6].set_index('job_posted_month_no')
df_percent
```

```python
fig,ax = plt.subplots(1,1)
sns.lineplot(df_percent , dashes=False , palette= 'tab10' )
sns.set_theme(style='ticks')
sns.despine() # removes border

plt.title('Trending Top Skills for Data Analysts in the US')
plt.ylabel('Percentage Likelihood in Job Posting')
plt.xlabel('2023')
plt.legend().remove()
from matplotlib.ticker import FuncFormatter

ax.yaxis.set_major_formatter(FuncFormatter(lambda x, pos: f'{x:.0f}%'))


for i in range(5):
    plt.text(12.2, df_percent.iloc[-1, i], df_percent.columns[i], color='black')

plt.show()
```
